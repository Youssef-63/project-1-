```python
X = data.iloc[:,0]
Y = data.iloc[:,1]
```

```python
def loss(y, y_predicted):
    n = y.count()
    loss = np.sum((y - y_predicted)**2) / n
    return loss
```

```python
def gradient_descent(X, Y, iterations = 1000, learning_rate= 0.1):
    w = 0
    b = 0
    n = len(X)
    for i in range(iterations):
        Y_predicted = (w * X) + b
        cost = loss(Y , Y_predicted)
        w_derv = -(2/n) * sum(X * (Y - Y_predicted))
        b_derv = -(2/n) * sum(Y - Y_predicted)
        w = w - (learning_rate * w_derv)
        b = b - (learning_rate * b_derv)
        print(f" Iteration {i}: l'erreur : {cost},  w : {w}, b : {b}")
    return w,b
  
```

```python
w,b = gradient_descent(X,Y,iterations=1000,learning_rate=0.01)
```

```python
Y_predicted = w*X+b
plt.scatter(X,Y)
plt.plot([min(X), max(X)], [min(Y_predicted), max(Y_predicted)] , color='red')
plt.show()
```
